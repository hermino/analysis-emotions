```python
import glob
import random
import cv2 as cv
import numpy as np
import pandas as pd
from math import floor
from shutil import copyfile
```

```python
emotions = ["Anger", "Disgust", "Fear", "Happiness", "Sadness", "Surprise"]
```

```python
features = []
labels = []

for emotion in emotions:
    imagens = glob.glob("./emotionss/{}/*".format(emotion))
    for img in imagens:
        matriz = cv.imread(img)
        gray = cv.cvtColor(matriz, cv.COLOR_BGR2GRAY)
        features.append(gray.reshape((350*350)))
        labels.append(emotions.index(emotion))
```

```python
df_imagens = pd.DataFrame(features)
df_imagens.head()
```

```python
df_imagens['label'] = labels
df_imagens.head()
```

```python
df_imagens.to_csv('./df_imagens.csv', index=False)
```
